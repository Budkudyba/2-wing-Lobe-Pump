```python
# we try to make a spirograph

import numpy as np
import matplotlib.pyplot as plt
```

```python
#basic form of a circle upon an circle
#x(θ)=(R−r)cos(θ)+rpcos(R+rrθ)
#y(θ)=(R−r)sin(θ)−rpsin(R+rrθ)

# pos(x or y) = large circle +/- smaller circle
```

```python
gradius = np.arange(0, 1, 0.000001) #angle plot resolution
theta = 2 * np.pi * gradius #convert to polar

r = 1 #rolling circle
R = 4 #exterior circle
p = 1 #position of the mark left behind within the "rolling" circle
fig, graph = plt.subplots() 

#inside spiro
ax = (R-r)*np.cos(theta + np.pi) - r*p*np.cos(1*(theta*((R-r))))
ay = (R-r)*np.sin(theta + np.pi) + r*p*np.sin(1*(theta*((R-r))))
graph.plot(ax, ay, color='blue')

#outside spiro
r = 1 #rolling circle
R = 4 #exterior circle
p = 1 #position
ax = (R+r)*np.cos(theta + np.pi) + r*p*np.cos(theta*((R+r)/r))
ay = (R+r)*np.sin(theta + np.pi) + r*p*np.sin(theta*((R+r)/r))
graph.plot(ax, ay, color='orange')

#can reduce resolution
#gradius = np.arange(0, 1, 0.001) #angle plot resolution
#theta = 2 * np.pi * gradius #convert to polar

# inner rotation guide circle
ax = r*np.cos(theta) - R + r
ay = r*np.sin(theta) + 0
graph.plot(ax, ay, color='red')

# large circle 
ax = R*np.sin(theta)
ay = R*np.cos(theta)
graph.plot(ax, ay, color='green')

#outer rotation guide circle
ax = r*np.cos(theta) - R - r
ay = r*np.sin(theta) + 0
graph.plot(ax, ay, color='purple')

#maintain equal axis spacing
graph.axis('equal')
#graph.grid(True)
#graph = plt.plot()

#output and save a svg to extude as lobes
fig.set_size_inches(10, 10)
plt.show()

#fig.savefig("lobe001.svg", dpi=92)
```
